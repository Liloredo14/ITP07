```python
class Candidate:
  # Class attribute to store all votes
  votes = []

  def __init__(self, name, position):
    self.name = name
    self.position = position

  def setVotes(self, vote):
    self.votes.append(vote)  # Append vote to candidate's list

  def getVotes(self):
    return sum(self.votes)  # Return total votes for the candidate

  # Interface method (assuming you want to display candidate info)
  def display(self):
    print(f"Candidate: {self.name} ({self.position})")

# Create candidate objects
candi1 = Candidate("Alice", "President")
candi2 = Candidate("Bob", "Secretary")

# Simulate voting (without user input for now)
candi1.setVotes(10)
candi1.setVotes(5)
candi2.setVotes(8)
candi2.setVotes(12)

# Get total votes for each candidate
total_votes_candi1 = candi1.getVotes()
total_votes_candi2 = candi2.getVotes()

# Print the results
candi1.display()
print(f"Total Votes: {total_votes_candi1}")
candi2.display()
print(f"Total Votes: {total_votes_candi2}")
```
